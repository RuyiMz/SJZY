```python
import pandas as pd

drug_sensitivity = pd.read_csv(r"CCLE\CCLE_NP24.2009_Drug_data_2015.02.24.csv")
L1000_CCLE_drug = pd.read_table(r"L1000_PhaseI_CCLE_druginfo.txt")
shared_pert = pd.read_table(r"LINCS-L1000\PhaseI\L1000_PhaseI_CCLE_sharedpert.txt")
#Create the file to store the matching results to the file:L1000_PhaseI_CCLE_result.txt
result_file = open(r"L1000_PhaseI_CCLE_result.txt","w",encoding="utf-8")
result_file.write("sig_id\tpert_id\tpert_iname\tpert_type\tcell_id\tpert_dose\tpert_dose_unit\tpert_idose\tpert_itime\tL1000_is_touchstone\tinchi_key\tpubchem_cid\tCCLE_Target\tDoses (uM)\tEC50 (uM)\tIC50 (uM)\tAmax\tActArea\tActArea_std\tIsEffective\n")

#Standardized active area
drug_sensitivity['ActArea_std'] = (drug_sensitivity['ActArea'] - drug_sensitivity['ActArea'].mean()) / drug_sensitivity['ActArea'].std()
#Determine whether the drug is effective
drug_sensitivity['IsEffective'] = 0
for ds_index in range(len(drug_sensitivity)):
    if drug_sensitivity.at[ds_index,'ActArea_std'] > 0.8:
        drug_sensitivity.at[ds_index,'IsEffective'] = 1
    if drug_sensitivity.at[ds_index,'ActArea_std'] < -0.8:
        drug_sensitivity.at[ds_index,'IsEffective'] = 0
    if (drug_sensitivity.at[ds_index,'ActArea_std'] >= -0.8) and (drug_sensitivity.at[ds_index,'ActArea_std'] <= 0.8):
        drug_sensitivity.at[ds_index,'IsEffective'] = -1
#Process the cell line data
drug_sensitivity['Cell_Line'] = ""
for cl_index in range(len(drug_sensitivity)):
    drug_sensitivity.at[cl_index,'Cell_Line'] = drug_sensitivity.at[cl_index,'Primary Cell Line Name'].strip().replace("-","").replace(" ","")
L1000_CCLE = pd.merge(drug_sensitivity,L1000_CCLE_drug,left_on='Compound',right_on='CCLE_compound_name',how="left")
#Delete drug information with standard deviation between -0.8 and 0.8
L1000_CCLE_filter = L1000_CCLE[L1000_CCLE['IsEffective'] != -1]
L1000_CCLE_filter = L1000_CCLE_filter.reset_index(drop=True)

counter = 0
for shared_pert_index in range(len(shared_pert)):
    for L1000_CCLE_index in range(len(L1000_CCLE_filter)):
        if (shared_pert.at[shared_pert_index,'pert_id'] == L1000_CCLE_filter.at[L1000_CCLE_index,'L1000_pert_id']) and (shared_pert.at[shared_pert_index,'cell_id'] == L1000_CCLE_filter.at[L1000_CCLE_index,'Cell_Line']):
            result_file.write(str(shared_pert.at[shared_pert_index,'sig_id'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_id'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_iname'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_type'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'cell_id'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_dose'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_dose_unit'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_idose'])+"\t")
            result_file.write(str(shared_pert.at[shared_pert_index,'pert_itime'])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,'L1000_is_touchstone'])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,'inchi_key'])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,'pubchem_cid'])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,'Target'])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"Doses (uM)"])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"EC50 (uM)"])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"IC50 (uM)"])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"Amax"])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"ActArea"])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"ActArea_std"])+"\t")
            result_file.write(str(L1000_CCLE_filter.at[L1000_CCLE_index,"IsEffective"])+"\n")
            counter = counter + 1
    print(str(shared_pert_index)+":"+str(counter))
            
result_file.close()
```
